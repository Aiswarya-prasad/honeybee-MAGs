Compare functional gene content across samples of different host species to identify functions that may be enriched or missing in certain host species and compare the extent to which functions are shared. Some of the first questions to ask are,
+ Is the gene catalog sufficiently representative?
+ Cumulative curve genes detected per host species?
+ Cumulative curves of KOs per host species?
+ Are certain KOs overrepresented in either?
+ What is the extent of sharing KOs?

To check if the gene catalog sufficiently represents the samples, we should consider the number of genes detected in each sample and then the cumulative curve across samples and also this for cd-hit clusters and KOs. For this we consider a gene / cluster / KO detected in a sample if 1 or more genes of the gene catalog corresponding to it is covered >= 1 and with the breadth > 50%

```python
samples = [x.split('/')[-1].split('_mapped')[0] for x in glob.glob('results/08_gene_content/01_profiling/*_mapped.coverage')]
for sample in samples:
    input_f = f'results/08_gene_content/01_profiling/{sample}_mapped.coverage'
    output_f = f'results/08_gene_content/01_profiling/{sample}_mapped.coverage.filtered'
    if os.path.isfile(output_f):
        print(f'{output_f} exists')
    else:
        # use awk to filter the coverage file
        print(f'{output_f} does not exist')
        os.system(f"awk '$6 > 50' {input_f} > {output_f}")
```

```python
coverage_outputs_filt = {}
for sample in samples:
    try:
        coverage_output_filt = pd.read_csv(f'results/08_gene_content/01_profiling/{sample}_mapped.coverage.filtered', sep='\t')
        coverage_outputs_filt[sample] = coverage_output_filt
    except:
        pass
```

```python
genes_per_sample = {}
for sample in samples:
    genes_per_sample[sample] = len(coverage_outputs_filt[sample])
with open('results/figures/visualize_temp/genes_per_sample.tsv', 'w') as out_fh:
    out_fh.write('sample\tgenes\n')
    for sample in genes_per_sample:
        out_fh.write(f'{sample}\t{genes_per_sample[sample]}\n')
```
